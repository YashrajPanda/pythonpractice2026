python supports 3 categories of data types -

    1.Basic Types : (integer, float, complex, boolean and string)
    2.Container Types : (List, Tuples, Sets and Dictionary)
    3.User-defined Types - Class

```python
# integer
print(3)
print(1e309)
```

```python
# float
print(3.14)
print(1e309)
```

```python
# boolean
print(True)
print(False)
```

```python
# complex
print(1 + 2j)
```

```python
# String
print("Kolkata")
print('Kolkata')
print("""Kolkata""")
```

```python
# List
print([1, 2, 3])
```

```python
#Typle
print((1, 2, 3))
```

```python
#Sets
print({1, 2, 3})
```

```python
# dictionary
print({"name": "Yashraj", "age": 21, "Gender": "Male"})
```
