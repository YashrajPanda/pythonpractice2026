Operators

Operators are used to perform operations on variables and values. Python has the following operation.

1. Arithmetic Operators

2. Comparison Operators

3. Logical Operators

4. Bitwise operators

5. Assignment Operators

6. Identity Operators

7. Membership Operators


1. Arithmetic Operators

```python
x = 5
y = 2

print(x + y)
```

```python
print(x - y)
print(x * y)
print(x / y) # true division
print(x // y) # floor division
print(x % y)
print(x ** y)
```

2. Comparison Operators

```python
print(x > y)
```

```python
print(x < y)
```

```python
print(x >= y)
```

```python
print(x <= y)
```

```python
print(x == y)
```

```python
print(x != y)
```

3. Logical Operators

6. Identity Operators : if they are in same memory location

```python
a = 3 
b = 3
print(a is b) # checks if a and b are the same object in memory
print(a == b) # checks if a and b have the same value
```

```python
a = "hello"
b = "hello"

print(a is b)
```

```python
a = [1, 2, 3]
b = [1, 2, 3]

print(a is b) # checks if a and b are the same object in memory
```

```python
a = "Hello-World"
b = "Hello-World"

print(a is b)
```

7. Membership Operators

```python
x = "Delhi"
print("D" in x)
```

```python
x = [1, 2, 3]
print(3 in x) # checks if 3 is an element of the list x
```

```python
x = (1, 2, 3)
print(3 in x) # checks if 3 is an element of the tuple x
```
